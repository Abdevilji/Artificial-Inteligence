```python
def stepf(x, t):
    return 1.0 if (x >= t) else 0.0

n=2
a=0.3
t=0.9
epochs=200
I = [[1, 1], [0, 0],[1, 0], [0, 1]]
T = [1, 0, 0, 0]
W = [0.0, 0.0]
Err = [0 for _ in range(len(T))]
mErr = 1
Error = []

for i in range(len(I)):
    for j in range(len(I[i])):
        print(I[i][j], end=" ")
    print(T[i], end="\n")
print()

line = ""
epoch=0
while epoch<epochs and mErr>0.04:
    epoch+=1
    print('Epoch=', str(epoch), end="\t\t")
    line=""
    for i in range(len(T)):
        net=0
        if i>0: line=line+"\t"
        for j in range(n):
            #print(T[i])
            if j>0: line = line + " + "
            line = line + str(I[i][j])+'*'+str(W[j])
            net = net + I[i][j]*W[j]
        net = round(net, 4)
        line=line+" = "+str(net)+"\t"+str(stepf(net, t))+"\t"+str(Err[i])+""
        if T[i]!=stepf(net, t):
            Err[i] = round(T[i]-net, 4)
            for k in range(n):
                W[k] = round(W[k]+a*I[i][k]*Err[i], 4)
                line+="\tW["+str(k+1)+"]="+str(W[k])
        line+="\n"
    mErr = round(sum(Err)/len(Err), 4)
    Error.append(mErr)
    line=str(mErr)+"\n\t"+line
    print(line)
    
x = [i for i in range(1, len(Error)+1)]
from matplotlib import pyplot as plt
plt.plot(x, Error, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Error')
```
